Making Graphs

```python
import matplotlib.pyplot as plt

# Data
num_trees = [100, 300, 500, 1000]
accuracy_stick_vs_avatar = [0.49536, 0.49502, 0.49502, 0.49502]
accuracy_valid_vs_invalid = [0.49107, 0.49381, 0.49365, 0.49210]

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(num_trees, accuracy_stick_vs_avatar, width=30, label='Stick vs Avatar Cue')
plt.bar(num_trees, accuracy_valid_vs_invalid, width=20, label='Valid vs Invalid')

# Axes labels and title
plt.xlabel('Number of Trees')
plt.ylabel('Accuracy')
plt.title('Performance of Random Forest Model with Feature Extraction')

# X-axis ticks
plt.xticks(num_trees)
plt.ylim([0.48, 0.5])

# Legend
plt.legend()

# Display the graph
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
pre_process = ['Without Feature Selection', 'With Feature Selection']
evaluation = ['Subset 1', 'Subset 2']
accuracy_valid_vs_invalid = np.array([[0.50197, 0.50085],
                                      [0.50062, 0.50058]])
accuracy_stick_vs_avatar = np.array([[0.51582, 0.51532],
                                     [0.50512, 0.52215]])

# Plotting the data
plt.figure(figsize=(8, 6))
width = 0.35
x = np.arange(len(evaluation))

plt.bar(x - width/2, accuracy_valid_vs_invalid[:, 0], width, label='Valid vs Invalid')
plt.bar(x + width/2, accuracy_stick_vs_avatar[:, 0], width, label='Stick vs Avatar Cue')

# Axes labels and title
plt.xlabel('Evaluation')
plt.ylabel('Accuracy')
plt.title('Performance of Random Forest Model')

# X-axis ticks and labels
plt.xticks(x, evaluation)

# Y-axis range
plt.ylim([0.45, 0.55])

# Legend
plt.legend()

# Display the graph
plt.show()
```
