```python

expdir = r"C:/Users/0412m/OneDrive/Bureaublad/Thesis/Group1_feature"

files = os.listdir(expdir)
#files = [f for f in files if os.path.isfile(os.path.join(expdir, f))]
print(files)

# here the data is loaded and separated into the conditions
nsubj = len(files)

# Stick data Alldata
# All data regardless of accuracy
freqStickAllDbBL = [None] * nsubj
freqStickCongAllDbBL = [None] * nsubj
freqStickIncongAllDbBL = [None] * nsubj

# Gaze data Alldata
# All data regardless of accuracy
freqPeopleAllDbBL = [None] * nsubj
freqPeopleCongAllDbBL = [None] * nsubj
freqPeopleIncongAllDbBL = [None] * nsubj

valid_channels = ["Fp1", "Fpz", "Fp2", "F7", "F3", "Fz", "F4", "F8", "FC1", 
                  "FC2", "FC6", "T7", "C3", "Cz", "C4", "T8", "CP5", "CP1", 
                  "CP2", "CP6", "P7", "P3", "Pz", "P4", "P8", "POz"]

#df_freqStickAllDbBL = pd.DataFrame()
df_freq_StickCongAll = pd.DataFrame()
df_freq_StickIncongAll = pd.DataFrame()
df_freq_PeopleCongAll = pd.DataFrame()
df_freq_PeopleIncongAll = pd.DataFrame()


for subnum, file in enumerate(files):
    subjdir = os.path.join(expdir, file)
    CurrentDir = os.path.join(subjdir, 'ProcessedData')
    out_dir = os.path.join(subjdir, 'results')
    if not os.path.exists(out_dir):
        os.mkdir(out_dir)

    datafileC = os.path.join(CurrentDir, 'data_ica.mat')
    
    print(datafileC)

    data_freqShifttoProbe = scipy.io.loadmat(datafileC)
    amount_trials = (len(data_freqShifttoProbe['trial'][0]))
    print(len(data_freqShifttoProbe['trial'][0]))

    valid_channel_indices = np.where(np.isin(data_freqShifttoProbe['label'], valid_channels))[0]

    if int(amount_trials) < 224:
      print("the amount of trials of participant " + str(subnum) + " is less than 224")
      continue 

  
##############
    print("starting with All Stick Cong Data")
    # Stick All Stick Cong data
    idx = np.where(data_freqShifttoProbe['trialinfo'][:, 0] == 1)[0]
    freq_StickCongAll = data_freqShifttoProbe['trial'][0][idx]
    
    maxi_StickCongAll = 0 
    for i in range(0, len(freq_StickCongAll)):
      if len(freq_StickCongAll[i][0]) > maxi_StickCongAll:
         maxi_StickCongAll = len(freq_StickCongAll[i][0])
    
    if maxi_StickCongAll> 15000: 
      print("the data of participant: "+ str(subnum) +" has too many values: " + str(maxi_StickCongAll) + "." )
      continue 

    #make blank_array 
    blank_array_StickCongAll = np.zeros((len(idx), len(valid_channels), maxi_StickCongAll))

    #put values in the array 
    for i in range(0, len(freq_StickCongAll)):
      for b in range(0, maxi_StickCongAll):
        for o, channel in enumerate(data_freqShifttoProbe['label']):
          if channel in valid_channels: 
            try: 
              blank_array_StickCongAll[i][o][b] = freq_StickCongAll[i][o][b]
            except IndexError: 
              pass 
    ##############
    # Stick All Stick Incong data
    idx = np.where(data_freqShifttoProbe['trialinfo'][:, 0] == 2)[0]
    freq_StickIncongAll = data_freqShifttoProbe['trial'][0][idx]

        
    maxi_StickIncongAll = 0 
    for i in range(0, len(freq_StickIncongAll)):
      if len(freq_StickIncongAll[i][0]) > maxi_StickIncongAll:
         maxi_StickIncongAll = len(freq_StickIncongAll[i][0])
    
    if maxi_StickIncongAll > 15000: 
      print("the data of participant: "+ str(subnum) +" has too many values: " + str(maxi_StickIncongAll) + "." )
      continue 

    #make blank_array 
    blank_array_StickIncongAll = np.zeros((len(idx), len(valid_channels), maxi_StickIncongAll))

    #put values in the array 
    for i in range(0, len(freq_StickIncongAll)):
      for b in range(0, maxi_StickIncongAll):
        for o, channel in enumerate(data_freqShifttoProbe['label']):
          if channel in valid_channels: 
            try: 
              blank_array_StickIncongAll[i][o][b] = freq_StickIncongAll[i][o][b]
            except IndexError:
              pass
##############
    ##############
    # People All People Cong data
    print("2/4 done")
    idx = np.where(data_freqShifttoProbe['trialinfo'][:, 0] == 3)[0]
    freq_PeopleCongAll = data_freqShifttoProbe['trial'][0][idx]

    maxi_PeopleCongAll = 0 
    for i in range(0, len(freq_PeopleCongAll)):
      if len(freq_PeopleCongAll[i][0]) > maxi_PeopleCongAll:
         maxi_PeopleCongAll = len(freq_PeopleCongAll[i][0])
    
    if maxi_PeopleCongAll > 15000: 
      print("the data of participant: "+ str(subnum) +" has too many values: " + str(maxi_PeopleCongAll) + "." )
      continue 

    #make blank_array 
    blank_array_PeopleCongAll = np.zeros((len(idx), len(valid_channels), maxi_PeopleCongAll))

    #put values in the array 
    for i in range(0, len(freq_PeopleCongAll)):
      for b in range(0, maxi_PeopleCongAll):
        for o, channel in enumerate(data_freqShifttoProbe['label']):
          if channel in valid_channels: 
      #  for o in range(0, len(freq_PeopleCongAll[0])):
            try: 
              blank_array_PeopleCongAll[i][o][b] = freq_PeopleCongAll[i][o][b]
            except IndexError:
              pass  
##############
    ##############
    # People All People Incong data
    idx = np.where(data_freqShifttoProbe['trialinfo'][:, 0] == 4)[0]
    freq_PeopleIncongAll = data_freqShifttoProbe['trial'][0][idx]

    maxi_PeopleIncongAll = 0 
    for i in range(0, len(freq_PeopleIncongAll)):
      if len(freq_PeopleIncongAll[i][0]) > maxi_PeopleIncongAll:
         maxi_PeopleIncongAll = len(freq_PeopleIncongAll[i][0])

    if maxi_PeopleIncongAll > 15000: 
      print("the data of participant: "+ str(subnum) +" has too many values: " + str(maxi_PeopleIncongAll) + "." )
      continue 

    #make blank_array 
    blank_array_PeopleIncongAll = np.zeros((len(idx), len(valid_channels), maxi_PeopleIncongAll))

    #put values in the array 
    for i in range(0, len(freq_PeopleIncongAll)):
      for b in range(0, maxi_PeopleIncongAll):
        for o, channel in enumerate(data_freqShifttoProbe['label']):
          if channel in valid_channels: 
            try: 
              blank_array_PeopleCongAll[i][o][b] = freq_PeopleIncongAll[i][o][b]
            except IndexError:
              pass  
##############


    # Freq analysis
    cfg = {}
    #cfg['channel'] = ['all']
    #cfg['method'] = 'wavelet'

    # mne -> epoch_data, sfreq, freqs, n_cycles=7.0, zero_mean=False, use_fft=True, decim=1, output='complex', n_jobs=None, verbose=None
    cfg['n_cycles'] = 3.0  # three cycles per time-window in steps of 50ms
    cfg['output'] = 'power'  # return the power-spectra
    #cfg['pad'] = 'nextpow2'
    cfg['freqs'] = np.arange(2, 31, 1)  # meaning 2-30Hz (for every 1Hz)
    cfg['verbose'] = 1

    # Stick All Stick data output = (n_chans, n_freqs, n_times)
    #freqStickAllDbBL[subnum] = mne.time_frequency.tfr_array_morlet(blank_array_StickAll, sfreq=512, **cfg)
        # Stick All Stick Cong data
    freqStickCongAllDbBL[subnum] = mne.time_frequency.tfr_array_morlet(blank_array_StickCongAll, sfreq=512, **cfg)

    # Stick All Stick Incong data
    freqStickIncongAllDbBL[subnum]= mne.time_frequency.tfr_array_morlet(blank_array_StickIncongAll, sfreq=512, **cfg)

    # People All people data
    #freqPeopleAllDbBL[subnum] = mne.time_frequency.tfr_array_morlet(blank_array_PeopleAll, sfreq=512, **cfg)

    # People All People Cong data
    freqPeopleCongAllDbBL[subnum] = mne.time_frequency.tfr_array_morlet(blank_array_PeopleCongAll, sfreq=512, **cfg)

    # People All People Incong data
    freqPeopleIncongAllDbBL[subnum] = mne.time_frequency.tfr_array_morlet(blank_array_PeopleCongAll, sfreq=512, **cfg)

print("done")
```

```python

# Calculate the total number of trials, channels, frequencies, and times
n_trials, n_chans, n_freqs, n_times = freqStickCongAllDbBL[0].shape
#n_times = int((freqStickCongAllDbBL[0].shape)[3]) 
#n_times = int(n_times/1)

# Create an empty 2-dimensional array
nsubj = len(freqStickCongAllDbBL)
data_2d = [] 

# Flatten the data and make a 2-dimensional array
for i, subj_data in enumerate(freqStickCongAllDbBL):
    subj_trials = subj_data.shape[0]
    flattened_data = subj_data.reshape(subj_trials, -1)

    print(flattened_data.shape[1])

    data_before = np.zeros((n_trials * nsubj +2, flattened_data.shape[1]))
    data_before[i * subj_trials: (i + 1) * subj_trials ] = flattened_data

    data_before = data_before.T

    data_before[:, :-2] = 1  
    data_before[:, :-1] = i 

    #data_before[i * subj_trials: (i + 1) * subj_trials, -2] = 1  # Add column with the value 1
    #data_before[i * subj_trials: (i + 1) * subj_trials, -1] = i  # Add subject information as the last column

    print(data_before.shape)

    if len(data_2d) == 0:
        data_2d = pd.DataFrame(data_before)
    else:
        df = pd.DataFrame(data_before)
        data_2d = pd.concat([data_2d, df], ignore_index = True)
        #data_2d = np.concatenate((data_2d, data_before), axis=0)


# Give the data column names
df = pd.DataFrame(data_2d, columns=[f'trial_{i}' for i in range(n_trials * nsubj)] + ['label', 'subject'])

# Save the data as a csv file
df.to_csv('StickCong_group1_feature_realtrainset.csv', index=False)
```
